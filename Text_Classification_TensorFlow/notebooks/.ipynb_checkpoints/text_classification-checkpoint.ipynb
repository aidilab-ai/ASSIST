```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import json
```

```python
data_path = '../data/'

# Load data
x_train = np.loadtxt(data_path + 'x_train.txt', dtype=str, delimiter='\n')
y_train = np.loadtxt(data_path + 'y_train.txt', dtype=str, delimiter='\n')
x_test = np.loadtxt(data_path + 'x_test.txt', dtype=str, delimiter='\n')
y_test = np.loadtxt(data_path + 'y_test.txt', dtype=str, delimiter='\n')

# For some reason, y_train contains 3 extra labels, so I remove them
y_train = y_train[:-3]
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# Load vocabulary
with open(data_path + 'vocabulary') as voc_file:
    voc = voc_file.read()
    voc = voc.replace('\'', '"')
    vocabulary = json.loads(voc)
```

```python
# Encode labels as binary vectors
y_train_one_hot = pd.get_dummies(y_train).values
y_test_one_hot = pd.get_dummies(y_test).values
```

```python
x_train[0]
```

```python
vocabulary.keys()
```

```python
from stop_words import get_stop_words

#vectorizer = CountVectorizer(max_features=200, min_df=5, max_df=0.7, stop_words=get_stop_words('italian'))  
vectorizer = TfidfVectorizer(max_features=200, min_df=5, max_df=0.7, stop_words=get_stop_words('italian'))

# Turn train and test sentences to bag of words
x_train_bow = vectorizer.fit_transform(x_train).toarray()
x_test_bow = vectorizer.fit_transform(x_test).toarray()
```

```python
model = MLPClassifier(hidden_layer_sizes=(8,))
model.fit(x_train_bow, y_train_one_hot)

y_test_pred = model.predict(x_test_bow)
```

```python
accuracy_score(y_test_one_hot, y_test_pred)
```
